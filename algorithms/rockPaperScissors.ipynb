/*
* Write a function that generates every sequence of throws a single
* player could throw over a three-round game of rock-paper-scissors.
*
* Your output should look something like:
*   [["rock", "rock", "rock"],
*    ["rock", "rock", "paper"],
*    ["rock", "rock", "scissors"],
*    ["rock", "paper", "rock"],
             ...etc...
     ]
*
* Extra credit:
*   - Make your function return answers for any number of rounds.
* Example:
* rockPaperScissors(5); // => [['rock', 'rock', 'rock', 'rock', 'rock'], etc...]
*
*/

```python
def rockPaperScissors(nbrRounds):
    
    throws = ['rock', 'paper', 'scissors']
    throwsSeq = []
    print('rounds number =', 3)
    
    if nbrRounds == 3:
        for throw1 in throws:
            for throw2 in throws:
                for throw3 in throws:
                    throwsSeq.append([throw1, throw2, throw3])
        print('every sequence', throwsSeq)
        print('there are', len(throwsSeq), 'sequences')
    else:
        print('works only for 3 rounds')
    
rockPaperScissors(3)
```

```python
def rockPaperScissors(nbrRounds):
    
    throws = ['rock', 'paper', 'scissors']
    throwsSeq = []
    
    def recFunc(roundLeft, throwPlayed):
        
        if roundLeft == 0:
            throwsSeq.append(throwPlayed)
            
        else:
            for throw in throws:
                recFunc(roundLeft - 1, throwPlayed + [throw])  
    
    recFunc(nbrRounds, []) 
    print(throwsSeq, len(throwsSeq))
    
rockPaperScissors(3)
```
